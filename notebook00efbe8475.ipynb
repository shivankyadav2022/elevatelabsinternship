Aligning with the sub-tasks given in task 1, we will perform following operations with the dataset : 
1. Identify and handle missing values
2. Remove duplicates
3. Strandradize text values in [Education] and [Marital_Status] columns
4. Convert date format of [Dt_Customer] to consistent date type (dd-mm-yyyy)
5. Rename column headers in snakecase (lowercase, and underscore for space)
6. Check and fix data types : [Dt_Customer] will be converted to datetime from Object

```python
#Check for missing values 
data.isnull().sum()
```

```python
#drop rows containing null values 
data_1= data.dropna().copy()

#check the data set after dropping null values 
data_1.isnull().sum()
```

```python
#Check for duplicate values in the dataset 
data_1.duplicated().sum()
```

The dataset contains no duplicate values. Lets now move towards standradization of text

```python
#Check For unique values in education column
data_1['Education'].unique()
```

```python
#Check for unique values in Marital_Status column
data_1['Marital_Status'].unique()
```

```python
# change the values to lower case in both columns 
data_1['Marital_Status'] = data_1['Marital_Status'].str.lower()
data_1['Education'] = data_1['Education'].str.lower()
```

```python
#Replace space with underscore 
data_1['Education'] = data_1['Education'].str.replace(' ', '_', regex=False)
```

```python
#check that values are standradidzed 
data_1['Education'].unique()
```

```python
# Change date format to datetime 
data_1['Dt_Customer'] = pd.to_datetime(data_1['Dt_Customer'],format='%d-%m-%Y')
```

```python
#Change column names to snakecase 
import re 
data_1.columns = [re.sub(r'\W+','_',col).strip('_').lower() for col in data_1.columns]
```

```python
#Check dataset first 5 rows 
data_1.head(5)
```

```python
#Write the cleaned dataset to CSV file 
data_1.to_csv('cleaned_data_task_1',index = False)
```
